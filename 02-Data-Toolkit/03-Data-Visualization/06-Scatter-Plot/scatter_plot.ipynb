# Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import and explore the dataset
```

```python
df = pd.read_csv('4_ThreeNum.csv')
df.head()
```

## Part 1

```python
fig, axs = plt.subplots(1, figsize=(15, 6))
ax1 = sns.scatterplot(x="gdpPercap",y="lifeExp", size="pop", hue="continent", sizes=(20, 1000), palette="muted",data=df)

handles, labels = ax1.get_legend_handles_labels()
labels[6] = 'Population (B)'
ax1.legend(handles[6:], labels[6:],loc="lower center", ncol=6, bbox_to_anchor=(0.4, -0.2), frameon=False)
sns.despine(ax = ax1,  top=True, right=True, left=True, bottom=True)
ax1.grid(color='#CCCCCC', linestyle='--', linewidth=0.6)
```

## Part 2

```python
fig, axs = plt.subplots(1, figsize=(15, 8))
ax1 = sns.scatterplot(x="gdpPercap",y="lifeExp", size="pop", hue="continent", sizes=(20, 1000), palette="muted",data=df)

handles, labels = ax1.get_legend_handles_labels()
labels[6] = 'Population (B)'
ax1.legend(handles[6:], labels[6:],loc="lower center", ncol=6, bbox_to_anchor=(0.4, -0.2), frameon=False)
sns.despine(ax = ax1,  top=True, right=True, left=True, bottom=True)
ax1.grid(color='#CCCCCC', linestyle='--', linewidth=0.6)
for index, row in df.iterrows():
    if row['gdpPercap']>5000 and row['lifeExp']<65:
        ax1.text(row['gdpPercap'], row['lifeExp'], row['country'])
```
