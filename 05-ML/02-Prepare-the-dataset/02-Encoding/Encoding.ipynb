# Encoding

üëá Import the work in progress dataset `cars_clean.csv` that you exported in the previous exercice.

```python
import pandas as pd
cars_clean = pd.read_csv('../data/cars_clean.csv')
cars_clean
```

## 1. `aspiration` & `enginelocation`

üëá Check the number of unique categories in the columns `aspiration` and `enginelocation`

```python
cars_clean['aspiration'].unique()
```

```python
cars_clean['enginelocation'].unique()
```

üëá Encode the columns using the appropriate encoding technique. Replace the original columns by the transformed values.


<details>
<summary>üí° Hint</summary>   
    
<br/>

[`LabelEncoder` documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html)
    
</details>



```python
from sklearn.preprocessing import LabelEncoder
#Instantiate the encoder
le = LabelEncoder()
#Fit the encoder on the required columns
cars_clean['aspiration_encoded'] = le.fit_transform(cars_clean['aspiration'])
cars_clean['enginelocation_encoded'] = le.fit_transform(cars_clean['enginelocation'])
```

## 2. `enginetype`

üëá Check the number of unique categories in the column `enginetype`

```python
type_list = cars_clean['enginetype'].unique()
type_list
```

üëá Encode `enginetype` using the appropriate technique. Replace the original column by the transformed values.


<details>
<summary>üí° Hint</summary>   
    
<br/>

[`OneHotEncoder` documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html)
    
</details>



```python
from sklearn.preprocessing import OneHotEncoder
le = OneHotEncoder(sparse = False)
# enginetype_oh = le.fit_transform(cars_clean[['enginetype']])
le.fit(cars_clean[['enginetype']])

category_list = le.categories_[0]

enginetype_oh = le.transform(cars_clean[['enginetype']])

for index, cat in enumerate(category_list):
    cars_clean[f'{cat}_encoded'] = enginetype_oh.T[index]
cars_clean
```

## 3. `cylindernumber`

üëá Check the number of unique categories in the column `cylindernumber`

```python
cars_clean['cylindernumber'].unique()
```

üëá Since the feature is ordinal, transform each category to its corresponding integer. Replace the original column by the transformed values.


<details>
<summary>üí° Hint</summary>   
    
<br/>

[`map` documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html)
    
</details>



```python
number_dict = {'four' : 4, 'six': 6, 'five': 5, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8}
```

```python
cars_clean['cylindernumber'] = cars_clean['cylindernumber'].map(lambda x: number_dict[x])
cars_clean
```

```python
cars_clean.to_csv(r'../data/cars_clean.csv', index = False)
```
